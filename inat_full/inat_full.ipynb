```python
combined_ymd = df['year'].astype(int).astype(str) + '-' + df['month'].astype(int).astype(str) + '-' + df['day'].astype(int).astype(str)
```

```python
# fix_zeroes = re.compile('\d{4}-')
def fix_zeroes(date):
    ymd = date.split('-')
    ans = ""
    ans += ymd[0] + '-'
    if len(ymd[1]) == 1:
        ans += "0"
    ans += ymd[1] + '-'
    if len(ymd[2]) == 1:
        ans += "0"
    ans += ymd[2]
    return ans
combined_ymd = combined_ymd.apply(fix_zeroes)
```

```python
eventDate_truncated = df.eventDate.apply(lambda x: str(x).split('T')[0])
```

```python
eventDate_truncated.equals(combined_ymd)
```
